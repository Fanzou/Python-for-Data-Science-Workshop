## Nested Conditions
Nested conditions are when we have a conditional statement(s) inside another conditional statement.

```python
x = 60


if x > 1: 
    print "x is More than 1"
    if x < 100:
        print "But less than 100"
    if x == 50:
        print "and is 50"
    if x != 50:
        print "and is not 50"

print "All done"
```

## If-Else Conditions (aka Two-Way Conditions)
If-Else conditions execute a block of code if a condition is true.

And if the condition is false, another block of code gets executed.

```python
Image(filename='images/s3_nestedcond.png') 
```

```python
# Do the above example
# Notice that the else-statement is not indented within the if-statement
x = 1
if x > 2:
    print "bigger"
else:
    print "not bigger"
print "All done"
```

### elif Condition (aka Multiway Condition)
Helps you understand whether a value falls within a specifc range.

i.e., Examine whether a number is higher than 10 but less than 100.


```python
x = 4

if  x < 2:
    print 'Small'
elif x < 10:
    print 'Medium'
else:
    print 'Large'
print 'All done'
```

```python
def get_cats_from_url(url):
    if len(url.split('/')) < 6:
        return "homepage"
    else:
        return url.split('/')[5]
```
