#### Print vs. Return

Print and Return are statements, not functions.

<b>Print</b> shows the user a string representing the output. It can help us understand how a program works, and is helpul for debugging.

<b>Return</b> gives us back a value and can be used to either store in a variable, or used as an argument which is passed into another function.

```python
def printingFun():
    print "I printed"
```

```python
printingFun()
```

```python
def returnFun():
    print "I printed again"
    return "I returned"
```

#### Break Statements

If you know ahead of time that your loop may go on infinitely, you can add a break statement to jump out of the loop.

```python
n = 10

while n > 0:
    print n
    n = n + 1
    if n > 100:
        break
```

### For-Loops and Lists

We use a for-loop to go through a list of items or through the contents of a file.

Constructing a loop involves:

1. Initializing one or more variables before the loop starts (this means you know ahead of time what variables you need in order to store the information you'll gather when you use the for-loop)

2. Peforming some computation or changing each item in the loop body

3. Looking at the resulting variables when the loop ends

<b>for</b> and <b>in</b> are used together in for-loops.

We will create a <b>list</b>, which is a sequence of numbers, characters, or a mix of both. 
Our list will consist of numbers only. 

We can use lists to perform a for-loop, which will modify the character in some way.


```python
count = 0
sampleList = [2, 3, 5, 10, 100]

for eachnum in sampleList:
    count = count + eachnum
    print count
```

We can also add the list directly in the for-loop. We get the same results.

```python
count = 0

for eachnum in [2, 3, 5, 10, 100]:
    count = count + eachnum
    print count
```

#### Making this into a function

Note that we need two arguments in the function:

<b>varList</b>: the variable containing the list of numbers

<b>singleVar</b>: contains a single value, which will get subtracted from the list of numbers in <b>varList</b>.

```python
y = [100, 50, 25, 10]
x = 1000


def subfunc(varList, singleVar):
    for eachnum in varList:
        singleVar = singleVar - eachnum
        print singleVar
    
```

```python
subfunc(y, x)
```
